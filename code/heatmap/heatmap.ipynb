```python
import folium
from folium import plugins
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from folium.plugins import MarkerCluster #import necessary items

%matplotlib inline
```

```python
df=pd.read_csv("sfpd_dispatch_data_subset.csv")
locs= pd.DataFrame(df['latitude'])
locs['longitude']=df['longitude']
locs.dropna(inplace=True)
locs2=locs.head(500) #using subset of data due to rendering limitations
locs2.head()
```

```python
m = folium.Map([37.774442, -122.5046], zoom_start=11) #check map
```

```python
m
```

```python
for index, row in locs2.iterrows():
    folium.CircleMarker([row['latitude'], row['longitude']],
                        radius=15,color="#26939E",
                        fill_color="#FFFFFF", fill=True
                       ).add_to(m) #loop through to map points
```

```python
folium.TileLayer('cartodbpositron').add_to(m)
```
