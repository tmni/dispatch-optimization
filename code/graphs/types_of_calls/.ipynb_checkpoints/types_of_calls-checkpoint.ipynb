```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.cm as cm
```

```python
df=pd.read_csv("sfpd_dispatch_data_subset.csv")
```

```python
#create a series of how many calls of each type
s=df.call_type.value_counts()
s.size #17 labels
```

```python
fig = plt.figure()
ax = fig.add_subplot(111)
fig.subplots_adjust(top=0.85)
ax.set_xlabel('amount of calls')
ax.set_ylabel('type of call')
x = np.arange(10)
ys = [i+x+(i*x)**2 for i in range(10)]
colors = cm.RdGy((np.linspace(0, 1, len(ys))))
s.plot(kind='barh', color=colors, figsize=(13,10))
```
