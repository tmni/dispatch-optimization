```python
import pandas as pd
import matplotlib.pyplot as plt
import datetime
from pandas.tools.plotting import table
```

```python
df=pd.read_csv("sfpd_dispatch_data_subset.csv") #create dataframe
```

```python
df.dispatch_timestamp=pd.to_datetime(df.dispatch_timestamp)
df.on_scene_timestamp=pd.to_datetime(df.on_scene_timestamp) 
```

```python
times= pd.DataFrame(df['on_scene_timestamp']-df['dispatch_timestamp']) #find length of time using time delta
```

```python
times['address'] = df['address'] 
```

```python
times.dropna(inplace=True)#remove Nat values
```

```python
times['minutes'] = (times[0].dt.total_seconds()/60) #convert timedelta to minutes
```

```python
avg_times=times.groupby('address', as_index=False)['minutes'].mean()#find avg by group
```

```python
avg_times.sort_values('minutes', inplace=True, ascending=False)
```
